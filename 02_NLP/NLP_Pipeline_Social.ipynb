```python
# Step 2
# Extract Spacy attributes
def step_2(df, irrelevant):
    df['NLP_relevant'] = df['NLP'].map(lambda x:[n for n in x if not n.is_stop if n.text not in irrelevant])
    df['Tokens'] = df['NLP_relevant'].map(lambda x:[n.text for n in x])
    df['Lemmas'] = df['NLP_relevant'].map(lambda x:[n.lemma_ for n in x])
    df['POSTags'] = df['NLP_relevant'].map(lambda x:[n.pos_ for n in x])
    df['Dependency'] = df['NLP_relevant'].map(lambda x:[n.dep_ for n in x])
    df['Entity'] = df['NLP_relevant'].map(lambda x:[n.ent_type_ for n in x])
    return df
```

```python
def NLPipe(df):
    
    #df['Text'] = df['Text'].str.replace('[^\w\s]', '')
    docs = list(nlp.pipe(df['Text']))
    tokens = []
    lemma = []
    pos = []
    dep = []
    ent = []
    irrelevant = ['Air','france','flew','flight', 'fly', ' ', 'china', 'emirates','singapore','lufthansa','air', 'canada']
    
    for doc in docs:
        if doc.is_parsed:
            tokens.append([n.text for n in doc if not n.is_stop if n.text not in irrelevant])
            lemma.append([n.lemma_ for n in doc if not n.is_stop if n.text not in irrelevant])
            pos.append([n.pos_ for n in doc if not n.is_stop if n.text not in irrelevant])
            dep.append([n.dep_ for n in doc if not n.is_stop if n.text not in irrelevant])
            ent.append([n.ent_type_ for n in doc if not n.is_stop if n.text not in irrelevant])
        else:
            # We want to make sure that the lists of parsed results have the
            # same number of entries of the original Dataframe, so add some blanks in case the parse fails
            tokens.append(None)
            lemma.append(None)
            pos.append(None)
            dep.append(None)
            ent.append(None)
    
    df['Tokens'] = tokens
    df['Lemmas'] = lemma
    df['POSTags'] = pos
    df['Dependency'] = dep
    df['Entity'] = ent
    #list_flat = functools.reduce(operator.iconcat, df['Tokens'],[])
    #top100 = pd.Series(list_flat).value_counts()[:n]
    return df
```

df = NLPipe(df)

```python
# Load the library with the CountVectorizer method
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# Helper function
def plot_10_most_common_words(count_data, count_vectorizer):
    import matplotlib.pyplot as plt
    words = count_vectorizer.get_feature_names()
    total_counts = np.zeros(len(words))
    for t in count_data:
        total_counts+=t.toarray()[0]
    
    count_dict = (zip(words, total_counts))
    count_dict = sorted(count_dict, key=lambda x:x[1], reverse=True)[0:50]
    words = [w[0] for w in count_dict]
    counts = [w[1] for w in count_dict]
    x_pos = np.arange(len(words)) 
    
    plt.figure(2, figsize=(15, 15/1.6180))
    plt.subplot(title='10 most common words')
    sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 2.5})
    sns.barplot(x_pos, counts, palette='husl')
    plt.xticks(x_pos, words, rotation=90) 
    plt.xlabel('words')
    plt.ylabel('counts')
    plt.show()

# Initialise the count vectorizer with the English stop words
count_vectorizer = CountVectorizer(stop_words='english')

# Fit and transform the processed titles
count_data = count_vectorizer.fit_transform(df['Lemmas'].map(lambda x: " ".join(x)))

# Visualise the 10 most common words
plot_10_most_common_words(count_data, count_vectorizer)
```

```python
# A helper function to get sentiment of a comment
def get_sentiment(text):
    return nlp(text)._.sentimenter['compound']

def average_score(l_sentences):
    if len(l_sentences) == 0:
        return np.nan
    else:
        l_results = [get_sentiment(text) for text in l_sentences]
        return np.mean(l_results)

#returns a list of sentences if contains a list of words
def find_sentence_if_l_words(txt, l_words):
    seps = ["? ", ". ", "! ", ", "]
    for sep in seps:
        txt = txt.replace(sep, '. ')
    l = [t for t in txt.split('. ') if any(x in t for x in l_words)]
    return l

#returns new features of list of sentence according to words related to a topic
def sentence_extraction_scoring(df_reviews, dict_topics, column_name='Text'):
    df = df_reviews.copy()
    for key, l_words in dict_topics.items():
        df[key + '_sentences'] = df.apply(lambda row: find_sentence_if_l_words(row[column_name], l_words), axis=1)
        df[key + '_score'] = df[key + '_sentences'].apply(average_score)
    return df

#returns new features of list of sentence according to words related to a topic
def sentence_extraction(df_reviews, dict_topics, column_name='Text'):
    df = df_reviews.copy()
    for key, l_words in dict_topics.items():
        df[key] = df.apply(lambda row: find_sentence_if_l_words(row[column_name], l_words), axis=1)
    return df
```
