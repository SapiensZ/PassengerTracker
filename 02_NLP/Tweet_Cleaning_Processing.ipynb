```python
# Removing " 'b "

df["Text"] = [x.strip("'b") for x in df.Text]
df['Text']
```

```python
# Removing RT

df["Text"] = [x.strip("RT ") for x in df.Text]
df['Text']
```

```python
# Remove @ tags and their account

df["Text"] = [re.sub(r'\w*@\w*', '', x) for x in df.Text]
df['Text']
```

```python
# Removing links from tweets

r = '(https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9]+\.[^\s]{2,}|www\.[a-zA-Z0-9]+\.[^\s]{2,})'
df["Text"] = df['Text'].str.replace(r, '')
df["Text"]
```

```python
# Remove punctuation

df['Text'] = df['Text'].str.replace('[^\w\s]', '')
df['Text']
```

```python
def NLPipe(df, n=100):
    
    df['Text'] = df['Text'].str.replace('[^\w\s]', '')
    docs = list(nlp.pipe(df['Text']))
    tokens = []
    lemma = []
    pos = []
    dep = []
    ent = []
    
    for doc in docs:
        if doc.is_parsed:
            tokens.append([n.text for n in doc if not n.is_stop])
            lemma.append([n.lemma_ for n in doc if not n.is_stop])
            pos.append([n.pos_ for n in doc if not n.is_stop])
            dep.append([n.dep_ for n in doc if not n.is_stop])
            ent.append([n.ent_type_ for n in doc if not n.is_stop])
        else:
            # We want to make sure that the lists of parsed results have the
            # same number of entries of the original Dataframe, so add some blanks in case the parse fails
            tokens.append(None)
            lemma.append(None)
            pos.append(None)
    
    df['Tokens'] = tokens
    df['Lemmas'] = lemma
    df['POSTags'] = pos
    df['Dependency'] = dep
    df['Entity'] = ent
    list_flat = functools.reduce(operator.iconcat, df['Tokens'],[])
    top100 = pd.Series(list_flat).value_counts()[:n]
    return top100
```

```python
def createWordcloud(wc):

    # Create and generate a word cloud image:
    wordcloud = WordCloud().generate(wc)

    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.figure(figsize=(10, 10))
    plt.show()
```

```python
def wordCloud(wc, image):
    
    mask = np.array(Image.open(image))
    image_colors = ImageColorGenerator(mask)
    disp = WordCloud(background_color="black", max_words = 100, width=400, height=400, mask=mask, random_state=1).generate(wc)
    # to recolour the image
    plt.figure(figsize = (10,10))
    plt.axis('off')
    plt.imshow(disp.recolor(color_func=image_colors))
```
